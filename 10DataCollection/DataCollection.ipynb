Data Collection

Data collection depends on the type of problem we want to solve.
Data can come from various sources such as Kaggle, SQL databases, devices, APIs, etc.

Sources of Data
1. Primary Data Source

Data collected first-hand for a specific problem or purpose.

Examples: surveys, interviews, sensors, experiments, manual observations.

2. Secondary Data Source

Data that already exists and is collected by someone else.

Examples: Kaggle datasets, government reports, research papers, company databases.

3. System-Generated Data

Data automatically produced by digital systems.

Examples: logs, app usage data, video feeds, IoT device outputs, synthetic data.

Sources

Databases – SQL (MySQL, PostgreSQL etc.), NoSQL databases (MongoDB etc.)- Sec

Public Datasets – Kaggle, UCI ML Repo etc.-Sec

APIs – Twitter API, Stock API, Weather API etc.-Sec

Web Scraping – HTML pages, blogs/articles, social media pages etc.-Sec

Sensors & Devices – IoT devices, health wearables etc.-Pri

Surveys & Forms – customer satisfaction, feedback etc.-Pri

System Generated – logs, video feeds, synthetic data etc.-Sys
