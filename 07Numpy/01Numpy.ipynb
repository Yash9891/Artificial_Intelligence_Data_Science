Definition:
NumPy (Numerical Python) is a Python library used for fast mathematical and scientific computing, providing powerful tools for working with large arrays and numerical data.


Fast Numerical Computing – NumPy is highly optimized, making computations much faster than pure Python.

N-Dimensional Arrays – It supports multi-dimensional arrays that can be up to 100× faster than Python lists.

Homogeneous & C-Optimized – NumPy arrays store same-type data and are implemented in C, giving high performance and efficient memory use.

```python
import numpy as np
import time
```

```python
arr=np.array([1,2,3,4,5,7,8,9,20])
print(arr)
print(type(arr))
```

```python
size=10_000_000
# calculate time
py_list=list(range(size))
start=time.time()
sq_list=[x**2 for x in py_list]
end=time.time()
print(f"Python list time ={end-start} seconds")

#calculate time
np_arr=np.array(py_list)
start=time.time()
# Vectorization- we not need to use loop
sq_array=np_arr**2
end=time.time()
print(f"Numpy array time ={end-start} seconds")
```

```python
# memory
import sys

print(f"Pyhton list size ={ sys.getsizeof(py_list)*len(py_list)} bytes")# it does not store continous memory
print(f"Numpy array size ={ np_arr.nbytes} bytes")# store data in continous memory block
```
