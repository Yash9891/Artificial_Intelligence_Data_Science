Aggrate functions

```python
import numpy as np

arr=np.array([1,2,3,4,5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))

print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
```

Power funs

```python
arr=np.array([1,2,3,4,5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr,2))
```

Log

```python
arr=np.array([1,2,3,4,5])

print(np.log(arr))# natural log - base e
# log base 10, log base 2
print(np.log2(arr))
print(np.exp(arr)) # e ki power x
```

Roundoff funs

```python
print(np.round(3.14,1))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(3.14))

```
