Pie Chart

A pie chart is a circular chart divied into slices to show relative prportions of a whole
- Each slice- A category
- Size of slice- Value of that category relative to total

```python
import matplotlib.pyplot  as plt
```

```python
#pie chart
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
# plt.style.use("dark_background")
plt.style.use("default")
explode=[0,0,0,0.1,0]
plt.pie(amounts,labels=expenses,autopct="%1.1f%%",
        wedgeprops={
            "edgecolor":"black",
            "linewidth":0.8,
            "linestyle":"-"
        },
        explode=explode,
        startangle=90,
        shadow=True
        )
plt.show()

```
