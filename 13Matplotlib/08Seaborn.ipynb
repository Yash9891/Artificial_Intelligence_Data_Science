Seaborn is a python data visualization library based on matplotlib.
Seaborn is use as high level where we do not want to change every minor detial in the graph. Just want to show high level graph in a faster way.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()# set theme
```

```python
sns.get_dataset_names() # default data set
```

```python
tips=sns.load_dataset("tips")
print(type(tips))
print(tips.head())
```

```python
# relational plots- scatter, line

#bill-tip - scatter chart

sns.set_theme(style="darkgrid")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker", # hilight smokers,
    size="size",
    style="smoker"
)


```

```python
# line plot


sns.set_theme(style="darkgrid")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker", # hilight smokers,
    kind="line"

)


```

```python
x=[1,2,3,4,5,6,7,8]
y=[i*2 for i in x]
sns.relplot(

    x=x,
    y=x,
    kind="line"


)


```

```python
#line plot

# years-passenger travelling each year

# load dataset
flights = sns.load_dataset("flights")

# line plot
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)

# the sahde in line plot beacuse we consider 5 % margin for error
```

Categorical plot

Bar plot
Box plot

```python
# Male vs female tip

# print(tips.head)

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"# male-female
)
```

```python
# box plot

# see the outlier
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

Distribution plot

Histogram

```python
penguins=sns.load_dataset("penguins")
# print(penguins.head())

# number of penguins acc to penguin weight 
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

Matrix

Heatmaps-Plot rectangular data as color-encoded matrix

```python
# flights.head()
# Show heatmap for each year - each month- how many passengers

# make the data frame wide
flights_pivot=flights.pivot(index="month",columns="year",values="passengers")

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("Passengers heatmap")
```

```python
# Bet practise to use seaborn
fig,ax=plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    marker="o",
    errorbar=None
)

ax.set_title("Avg bill for each day")
```
