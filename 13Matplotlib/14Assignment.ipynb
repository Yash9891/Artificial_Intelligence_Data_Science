Q1. For the following dataset generated using NumPy:
np.random.normal(loc=70, scale=10, size=200)
• Plot a histogram of scores
• Draw a vertical line for the mean & try to interpret the results.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
scores = np.random.normal(loc=70, scale=10, size=200)

# Calculate mean
mean_score = np.mean(scores)

# Plot histogram
plt.hist(scores, bins=20)
plt.axvline(mean_score)
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Scores with Mean Line")

plt.show()
```

Q2. For the Seaborn penguins dataset, do the following:
• Plot flipper_length_mm vs body_mass_g
• Color by species

```python
import seaborn as sns

# Load dataset
penguins = sns.load_dataset("penguins")

# Scatter plot
sns.scatterplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

plt.title("Flipper Length vs Body Mass (Colored by Species)")
plt.show()
```

Q4. For the Seaborn tips dataset, do the following:
• Create a scatter plot of total_bill vs tip
• Color points by sex
• Use different markers for smoker
• Add a title and axis labels
• Analyze: Do smokers tend to tip differently?

```python
tips=sns.load_dataset("tips")
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker"
)
# Labels & title
plt.title("Total Bill vs Tip (Colored by Sex, Marker by Smoker)")
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

Q5. For the Seaborn tips dataset, do the following:
• Create a pivot table of average tip
◦ Rows: day
◦ Columns: time
• Plot the pivot table as a heatmap
• Annotate values
• Interpret which day–time combination performs best


```python
# Load dataset
tips = sns.load_dataset("tips")

# Create pivot table
pivot_table = tips.pivot_table(
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)

# Heatmap
sns.heatmap(
    pivot_table,
    annot=True,
    fmt=".2f"
)

plt.title("Average Tip by Day and Time")
plt.show()
```
