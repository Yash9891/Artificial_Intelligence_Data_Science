```python
# Real data
# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]  # in $

plt.plot(years,revenue)
plt.title("Oscar movie revenue each year")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.show()

```

Barcharts

A bar plot is a graph that displays the categorical data using rectangular bars

```python
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]  # in $

plt.bar(years,revenue, color="orange")
plt.title("Revenue in each year")
plt.xlabel("Year")
plt.ylabel("Revenue $")

# Create labels for bar
for i in range(len(years)):
    plt.text(years[i],revenue[i]+10,str(revenue[i]),ha="center")#(x - axis, y- axis, text to print, alignment)

# extend the limits of y axis
plt.ylim(0,max(revenue)+50)

```

Compare 2 set of data

```python
import numpy as np
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $

x=np.arange(len(years))# [0,1,2,3,4] index of eyars in x axis
width=0.4# by defaut bar width is 0.8
plt.bar(x-width/2,oscar_revenue,width,label="Oscar movies")# shifting to left
plt.bar(x+width/2,non_oscar_revenue, width,label="Non Oscar movies")# shifting to right

plt.title("Oscar VS Non Oscar")
plt.xlabel("years")
plt.ylabel("Revenue in $")
plt.legend()

# replace x values with years
plt.xticks(x,years)



```
