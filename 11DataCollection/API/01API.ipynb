API-using request
API is like an endpoint where we send request and get the response
-GET
-Pots
-Delete

```python
import requests

url="https://stephen-king-api.onrender.com/api/books"

res=requests.get(url)

#Status code 200- successfull
# res

jsondata=res.json()
jsondata
```

```python
# convesrt into pandas data frame

import pandas as pd

df=pd.json_normalize(
jsondata["data"])

df=df[["id","Title","Year","ISBN"]]
df
```
