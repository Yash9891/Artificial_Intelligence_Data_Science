```python
import pandas as pd
df=pd.read_csv("globalAirQuality.csv")
print(df)

df.describe()
```

```python
#selecting data columns wise

# print(df["city"])
# print(df["aqi"])
print(df[["aqi","city"]])
```

```python
#select rows-label based and index based

# print(df.loc[0])
# print(df.loc[0:3])# ending index is inluded with loc

df.iloc[0:3] # does not incluse last index
```

```python
# row and col based selection

# df.loc[0, "aqi"]
df.loc[0:4, ["aqi",'city']]

# df.iloc[0:4, ["aqi",'city']]# it will give error because iloc require only numerical values not string-we pass only numerical values into iloc

df.iloc[0:4,[2,3]]

```

```python
# aselect single scalar value

print(df.at[0,"city"])# return view not copy of data
print(df.iat[0,2])

# to create copy

cities=df["city"].copy()
cities[0]="Mumbai"
print(cities)
```

```python
#Filtering of data

# Aqi > 100

# df[df["aqi"]>100]
df[(df["aqi"]>100) & (df["temperature"]>30)]


# chaining
df[df["aqi"]>100][["city","aqi"]]
```

```python
aqi_data=df[(df["aqi"]>100) & (df["temperature"]>30)]

# print(aqi_data)

aqi_data.iloc[0]# with index
aqi_data.loc[6] # with label 6
```

Queru Method

Query return copy not view

```python
df.query("aqi>100") # we can pass a string like sql query

df.query("aqi>100 & temperature>30")[["city","aqi"]]

# If want to use any variable insire query method we can use @ for reference

val=100
df.query("aqi>@val")[["city","temperature"]]
```
