DataFrame- 2D labeled array

```python
import pandas as pd
info={
    "Name":["Adam","Padam","Reiv"],
    "Age":[12,34,54],
    "GPA":[9.2,8.3,7.9]

}

df=pd.DataFrame(info)

print(df, type(df))
print(df.index) # index vals
print(df.columns)# col vals
```

```python
df=pd.DataFrame([["Adam",34],["Bob",54],["Pop",67]], columns=["Name","Age"])
print(df)
```

```python
# create dataframe with numpy array

import numpy as np
arr=np.array([[1,2,3],[4,5,6]])

df=pd.DataFrame(arr,columns=["a","b","c"])
print(df)
```

```python
# CSV Data

df=pd.read_csv("employee.csv")
print(df)
```
