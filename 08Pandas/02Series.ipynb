Series

```python
import pandas as pd
```

```python
s=pd.Series([1,2,3,4,5,6])
print(s)
```

```python
print(type(s))
```

```python
#index
print(s[0])
print(s[2:])
```

```python
#custom label
s=pd.Series([23,56,34,65,12,"P"], index=["Adam", "Bob", "Kutt","Pop","lol","no"])
print(s)

print(s["Pop"]) # label based access

# print(s[0]) # index based

# need to use iloc i want to access with index 
print(s.iloc[0])


# Pandas include homogenous data type
```

```python

#vectorization - perform operations on each elm without loop

s1=pd.Series([1,2,3,4])
s2=pd.Series([1,2,3,4])

print(s1+s2)
```

```python
# Handel missing values with NaN
# Mutable values , immutable size

s1[0]=10
print(s1)
changed=s1.drop(0)# it will not affect the original series

print(s1)
print(changed)
```
