Cleaning Data

```python
import pandas as pd

df=pd.read_csv("raw_data.csv")
df
```

```python
# handel missing value

df.isnull() # isna()- return true if value is null

# count number of missing values in column
df.isnull().sum() 

# remove null values- rows
df.dropna()

# drop columns
df.dropna(axis=1)# cols

# fill null values with some val
df.fillna(10)

# replace age col
age_mean=df["age"].mean()
df["age"].fillna(age_mean)

cleaned_data=df.copy()
age_mean=cleaned_data["age"].mean()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)
cleaned_data


# fordward fill- copye the previouse value in next null and move forward

df.ffill()

# backward fill- coming from last and fill the prev val in null( bottom to up)

df.bfill()

```

Handel Dublicate values


```python
df.duplicated() # return true for duplicate values
df["country"].duplicated() # ckeck for column wise

df[["country","gender"]].duplicated()

# remove duplicates
df.drop_duplicates() # does not change original

# change the original
df2=df.copy()
df2.drop_duplicates(inplace=True)
```

```python
#data types

df.dtypes

# convert the data type
df2 = df.copy()
df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64")
print(df2.dtypes)
```
