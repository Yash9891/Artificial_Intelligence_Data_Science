Cleaning and structing data


```python
import json
```

```python
# Load the data

def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store.json")
print(data)
```

-1.Data is inconsistence-mixed type of ratings
-2.Missing data-age
-3.Duplicate data


```python
#CLean and structure the data

def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3, "four":4,"five":5}
    cleaned_data=[]
    unique_users=set()

    for user in data:
        #clean Ratings
          # ---------- Clean Ratings ----------
        raw_rating = user.get("rating", "")

        # convert rating safely to string BEFORE strip()
        raw_rating_str = str(raw_rating).strip().lower()
        
        if (raw_rating_str in text_to_num):
            raw_rating=text_to_num[raw_rating]
        # print(raw_rating)
        user["rating"]=raw_rating

        #Handel missing value
        raw_age=user.get("age") # get does not give error if the value not present
        if raw_age ==None or str(raw_age).strip() == "":
            user["age"]=None

        # Deduplication
        name_key=user['name'].strip()
        if name_key in unique_users:
            continue
        else:
            unique_users.add(name_key)
            cleaned_data.append(user)
    return cleaned_data
```

```python
clean_data(data)
```
