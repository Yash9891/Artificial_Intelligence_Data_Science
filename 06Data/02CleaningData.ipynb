Cleaning and structing data


```python
import json
```

```python
# Load the data

def load_data(filename):
    with open(filename,"r") as f:
        data=json.load(f)
    return data
```

```python
data=load_data("store.json")
print(data)
```

-1.Data is inconsistence-mixed type of ratings
-2.Missing data-age
-3.Duplicate data


```python
#CLean and structure the data

def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3, "four":4,"five":5}
    cleaned_data=[]
    unique_users=set()

    for user in data:
        #clean Ratings
          # ---------- Clean Ratings ----------
        raw_rating = user.get("rating", "")

        # convert rating safely to string BEFORE strip()
        raw_rating_str = str(raw_rating).strip().lower()
        
        if (raw_rating_str in text_to_num):
            raw_rating=text_to_num[raw_rating]
        # print(raw_rating)
        user["rating"]=raw_rating

        #Handel missing value
        raw_age=user.get("age") # get does not give error if the value not present
        if raw_age ==None or str(raw_age).strip() == "":
            user["age"]=None

        # Deduplication 
        name_key=user['name'].strip()
        if name_key in unique_users:
            continue
        else:
            unique_users.add(name_key)
            cleaned_data.append(user)
    return cleaned_data
```

Get meaningfull Insights


```python
# get the avg rating
data=clean_data(data)

def get_insights(data):
     total_rating=0
     for user in data:
          total_rating+=float(user["rating"])
     print(f"Avg Rating= {total_rating/len(data)}")

     #percentage of users with poor ratings
     poor_ratings=0
     for user in data:
          if(float(user["rating"])<3):
               poor_ratings+=1
     print(f" % of usres with poor rating= {poor_ratings/len(data)*100} %")
```

Build a product recommendation features(rule based, not ML)

```python
get_insights(data)
```

```python
# build recommendation feature

def get_recomm(data):
    recommendations=[]
    
    for user in data:
        curr_recom={}
        curr_recom["name"]=user["name"]
        if(float(user["rating"])>4):
            curr_recom["brand"]="Apple"
        else:
            curr_recom["brand"]="samsung"
        recommendations.append(curr_recom)
    return recommendations
```

```python
get_recomm(data)
```
