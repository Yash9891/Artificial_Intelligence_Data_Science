Iris flower Data set

```python
import pandas as pd 
# load data

flower=pd.read_csv("IRIS.csv")
flower.iloc[0:10]
print(flower.shape)
print(flower.dtypes)

#Summary statistics (mean, std, min, max)

#flower.describe()

num_cols = flower.select_dtypes(include="number").columns
flower[num_cols].agg(["min", "max", "std", "mean"])
```

```python
# B.Select only those rows where:
# • petal_length > 4.5
# • species = "Iris-virginica"

print(flower[flower["petal_length"]>4.5])
flower[flower["species"]=="Iris-virginica"]
```

```python
# C. Group by species and compute:
# • Average sepal_length
# • Maximum petal_width
# • Standard deviation of sepal_width

flower.groupby("species").agg(
    avg_length=("sepal_length", "mean")
)

flower.groupby("species")["petal_width"].max()

flower.groupby("species").agg(
    standar=("sepal_width","std")
)
```

```python
# D. Create a new column: "petal_ratio" = petal_length / petal_width
# • For each species, find the average petal_ratio.

flower["petal_ratio"]=flower["petal_length"]/flower["petal_width"]

flower.groupby("species")["petal_ratio"].mean()

#other way to add new col
# flower = flower.assign(petal_ratio = flower["petal_length"] / flower["petal_width"])

```
