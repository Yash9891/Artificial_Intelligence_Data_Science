Merging and Joining Data

```python
import pandas as pd 

df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "name": ["Adam", "Bob", "Charlie", "David"]
})

df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [2, 1, 4, 5],
    "amount": [250, 120, 300, 180]
})

df_customers
```

```python
df_orders
```

```python
# inner join - givin the matching rows from both the tables

pd.merge(df_customers,df_orders,on="customer_id")
```

```python
# Left join

pd.merge(df_customers,df_orders,on="customer_id",how="left")
```

```python
# right join

pd.merge(df_customers,df_orders,on="customer_id",how="right")
```

```python
# outer join

pd.merge(df_customers,df_orders,on="customer_id",how="outer")
```

Data concatenation

Stacking DataFrame on top of each other(rows) or side by side (colms)

```python
df1=pd.DataFrame({
    "id":[1,2,3],
    "name":["Charlii","parle","Kot"]
})

df2=pd.DataFrame({
    "id":[4,5,6],
    "name":["Kharlii","POp","Kowt"]
})

pd.concat([df1,df2]) # row wise
# overwite old labels
pd.concat([df1,df2], ignore_index=True)

#col wise
pd.concat([df1,df2], axis=1)
```
