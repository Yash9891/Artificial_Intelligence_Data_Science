Transforming Data(Feature Enginnering)

```python
import pandas as pd 

df=pd.read_csv("../08Pandas/raw_data.csv")

df2=df.copy()

# apply tax fun on income - applyfuns take funs
df2["tax"]=df2["income"].apply(lambda x:"20%" if x>=50000 else "10%")
df2
```

```python
#Map fun

gender_map = {"Male": "M", "Female": "F", "Unknown": "U"}

# Correct way â€” modify only the 'gender' column
df2["gender"] = df2["gender"].map(gender_map)

# Display the full DataFrame
df2  



```

```python
#Assing- we can create new col with assign fun

df2.assign(new_income=df2["income"]*1.1)
df2


```

```python
# replace old values

df2["country"]=df2["country"].replace("USA","US")
df2
```

```python
# Rname- rows or colmns

df2.columns=["ID","NAME","AGE","COUNTRY","GENDER","INCOME","TAX"]

# rename
df2.rename(columns={"INCOME": "Salary"})
df2

# rename row
df2.rename(index={1:"First"})
```

```python
# Sorting
df2=df2.fillna(50)
df2.sort_values("Salary")
sorted_df=df2.sort_values(["Salary","AGE"], ascending=False)# if salary is same then sorting will be based on age

# sort the index
sorted_df.sort_index()

# reset index - but keep the data as it is- crete new index col
sorted_df.reset_index()
# drop old index- do not preserve old index
sorted_df.reset_index(drop=True)

```

```python
# Ranking

sorted_df["Ranking"]=sorted_df["Salary"].rank(ascending=False,method='dense')#dense  to resolve tie- do not assignt avg
sorted_df["Ranking"]=sorted_df["Salary"].rank(ascending=False,method='max')#max  assign max value to same salary 
sorted_df
```

```python
# Reassign colmns

sorted_df[["ID","AGE","COUNTRY","GENDER","Salary","TAX","NAME","Ranking"]
]
sorted_df
```
